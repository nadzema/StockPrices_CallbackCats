```python
## Define function to grab stock list
def Key_Stats(gather="Total Debt/Equity (mrq)"):
    ## Grab list of directories for stocks - #print(stock_list) prints list of directories
    statspath = path+"/_KeyStats"
    stock_list = [x[0] for x in os.walk(statspath)]
    ticker_list = []
    ## create data frame for stock data from Yahoo Finance data
    stock_df = pd.DataFrame(columns = ["Date", 
                                       "Unix", 
                                       "Ticker", 
                                       "DE Ratio", 
                                       "Stock Price", 
                                       "Stock Percent Change", 
                                       "S&P 500 Date",
                                       "S&P 500 Value", 
                                       "SP500 Percent Change"])
    
    ## S&P 500 Data - import form csv and put into new data frame
    sp500_data = "sp500_alt.csv"
    sp500_df = pd.read_csv(sp500_data)
    sp500_df.head()
    
    
    ## Get each file name in each stock directory
    ## Use [1:] because we do not need the root directory(main folder)
    for each_directory in stock_list[1:25]:  
        each_file = os.listdir(each_directory)
        ticker = each_directory.split("\\")[1]
        ticker_list.append(ticker)
        
        ## Set new starting point to calculate percent change as the loop goes thru the tickers
        starting_stock_value = False
        starting_sp500_value = False
        
        if len(each_file) > 0:
            for file in each_file:
                ## Begin to pull data on the file
                date_stamp = datetime.strptime(file, "%Y%m%d%H%M%S.html")
                unix_time = time.mktime(date_stamp.timetuple())
                
                ## Open each stock file in _KeyStats folder
                full_file_path = each_directory + "/" + file
                
                source = open(full_file_path, "r").read()
                ## Need to format all the source str code the same, some HTML files have a break after :</td> and some do not
                source_format = source.replace('\n','')
                
                sp500_date = datetime.fromtimestamp(unix_time).strftime('%Y-%m-%d')
                row = sp500_df[sp500_df["Date"] == sp500_date]
                #sp500_value = sp500_df["Adj Close"]
                sp500_value = row["Adj Close"]
                
              #  sp500_df["Formatted Date"] = sp500_date
                ## Isolate the value we need by splitting the entire formatted HTML source code before and after the value we need
                ## Loop for Total Debt/Equity ratio
                if 'Total Debt/Equity (mrq):</td>' in source_format:
                    value = source_format.split(gather + ':</td><td class="yfnc_tabledata1">')[1].split('</td>')[0]
                elif 'Total Debt/Equity (mrq):</th>' in source_format:
                    value = source_format.split(gather + ':</th><td class="yfnc_tabledata1">')[1].split('</td>')[0]
                else: 
                    value = "No Data Available"  
                    
                ## New loop to pull Stock Price from Yahoo Finance Data    
                if '</small><big><b>' in source_format:
                    stock_price = source_format.split("</small><big><b>")[1].split("</b></big>")[0]
                elif '<span id="yfs_l10' in source_format:
                    stock_price = source_format.split('<span id="yfs_l10')[1].split("</span></b></big>")[0]
                elif '<span id="yfs_l84' in source_format:
                    stock_price = source_format.split('<span id="yfs_l84')[1].split("</span></span>")[0]
                else:
                    stock_price = "No Data Available"
                
                ##
      #          if not starting_stock_value:
      #              starting_stock_value = stock_price
      #          if not starting_sp500_value:
      #              starting_sp500_value = sp500_value
                    
                ## New minus Old divided by Old times 100
                ## New - Old / Old x 100
         #       stock_p_change = ((stock_price - starting_stock_value) / starting_stock_value) * 100
          #      sp500_p_change = ((sp500_value - starting_sp500_value) / starting_sp500_value) * 100
                
                ## append data to data frame 
                stock_df = stock_df.append({"Date": date_stamp, 
                                            "Unix": unix_time, 
                                            "Ticker": ticker, 
                                            "DE Ratio": value, 
                                            "Stock Price": stock_price, 
                               #             "Stock Percent Change": stock_p_change,
                                            "S&P 500 Date": sp500_date,                       
                                            "S&P 500 Value": sp500_value,
                               #             "S&P 500 Percent Change": sp500_p_change
                                           }, ignore_index = True)
                
                
    save = gather.replace(" ", "").replace(")", "").replace("(", "").replace("/", "") + (".csv")
    print(save)
    stock_df.to_csv(save)
                                            
```
